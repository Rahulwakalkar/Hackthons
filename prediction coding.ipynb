```python
# Load  libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot
from pandas.plotting import scatter_matrix
from sklearn.preprocessing import LabelEncoder
# Import Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
df=pd.read_csv("sample_user_data.csv") # Load data file

```

```python
df1 = df.drop(["Date"], axis=1) # drop date column
```

```python
count = df1.isna().sum() #find null values
print(count)
```

```python
#fill null values
l_na=['bounces', 'pageviews', 'timeOnSite']
for column in l_na:
    df1[column].fillna(df1[column].mode()[0], inplace=True)
```

```python
count = df1.isna().sum()
print(count)
```

```python
df1.dtypes #data types
```

```python
#to find uniques in columns
print("source\n", df1.source.unique())
print("\n medium \n", df1.medium.unique())
print(" \n campaign \n", df1.campaign.unique())
print(" \n deviceCategory \n", df1.deviceCategory.unique())
print(" \n operatingSystem \n", df1.operatingSystem.unique())
print(" \n mobileDeviceModel \n", df1.mobileDeviceModel.unique())
print(" \n city \n", df1.city.unique())
print(" \n ChannelGrouping \n", df1.ChannelGrouping.unique())

```

```python
len(df['transactions']) # length of the transactions
```

```python

# Binary encoding for transactions
for i in range(len(df['transactions'])):
    if pd.isnull(df['transactions'][i]):
        df1['transactions'][i]=0
    else:
        df1['transactions'][i]=1
        
df1.transactions.astype(int)
```

```python
df1['transactions'] = df1.transactions.astype(int) # convert datatype to int 
df1.head()
```

```python
df1.dtypes
```

```python
df1.transactions.unique()
```

```python
# drop  mobileDeviceModel as it contains only one value
df1 = df1.drop(['mobileDeviceModel'], axis=1)


devicecategory = pd.get_dummies(df1['deviceCategory'], drop_first=True)

df1 = pd.concat([df1, devicecategory], axis=1)
del df1['deviceCategory']
```

```python
df1.head(50)
```

```python
# Spiliting dependent and independent variables
dfx = df1.drop(['fullVisitorId', 'VisitNumber', 'VisitStartTime', 'totalTransactionRevenue', 'transactions'], axis=1)
dfx.head()

x = dfx.iloc[:,:].values
y = df1.iloc[:, 7].values
```

```python
print(y)
```

```python
dfx.dtypes
```

```python
#Label Encoding independent variables
le = LabelEncoder()
x[:,3] = le.fit_transform(x[:,3])
x[:,4] = le.fit_transform(x[:,4])
x[:,5] = le.fit_transform(x[:,5])
x[:,6] = le.fit_transform(x[:,6])
x[:,7] = le.fit_transform(x[:,7])
x[:,8] = le.fit_transform(x[:,8])
```

```python
x = x.astype(int)
y = y.astype(int)
```

```python
print(y)
```

```python
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split # Import train_test_split function

# split into train test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
## Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf=clf.fit(x_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(x_test)
```

```python
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation

print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```
